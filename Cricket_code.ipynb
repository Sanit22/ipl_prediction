For each batsmen, we find out strike rate, batting average, highest and lowest runs scored by the batsmen in an inning.

## Batsmen strike rate lifetime

Calculated as ((runs scored / balls faced) * 100)



```python
batsmen_strike = {}

# filtered_df = df.loc[(df['A'] > 10) & (df['B'] < 5), ['A', 'C']]

ball_data[ball_data['batter'] == 'JC Buttler']['batsman_run'].sum()
len(ball_data.loc[(ball_data['batter'] == 'JC Buttler') & (ball_data['extra_type'].isna())])

for batsmen in unique_batsmen:
  runs_scored = ball_data[ball_data['batter'] == batsmen]['batsman_run'].sum()
  balls_faced = len(ball_data.loc[(ball_data['batter'] == batsmen) & ((ball_data['extra_type'].isna()) | (ball_data['extra_type'] == 'byes') | (ball_data['extra_type'] == 'legbyes'))])
  strike_rate = (runs_scored / balls_faced) * 100
  strike_rate = round(strike_rate, 2)
  batsmen_strike[batsmen] = strike_rate

print(batsmen_strike)
```

# Runs metrics

## Average, max, min runs scored in that stadium in 6 overs in that particular inning

```python
# the number at the end indicates inning number
avg_runs_stadium_1 = {}
avg_runs_stadium_2 = {}
max_runs_stadium_1 = {}
max_runs_stadium_2 = {}
min_runs_stadium_1 = {}
min_runs_stadium_2 = {}


# avg_runs_batting_bowling_team = {}
# max_runs_batting_bowling_team = {}

stadiums = list(merged_df['Venue'].unique())

teams = list(merged_df['BattingTeam'].unique())

for stadium in stadiums:
  rows_1 = merged_df.loc[(merged_df['Venue'] == stadium) & (merged_df['innings'] == 1)]
  sum_1 = rows_1['total_run'].sum()
  avg_1 = sum_1 // len(rows_1)
  max_1 = rows_1['total_run'].max()
  min_1 = rows_1['total_run'].min()

  rows_2 = merged_df.loc[(merged_df['Venue'] == stadium) & (merged_df['innings'] == 2)]
  sum_2 = rows_2['total_run'].sum()
  avg_2 = sum_2 // len(rows_2)
  max_2 = rows_2['total_run'].max()
  min_2 = rows_2['total_run'].min()

  avg_runs_stadium_1[stadium] = avg_1
  avg_runs_stadium_2[stadium] = avg_2
  max_runs_stadium_1[stadium] = max_1
  max_runs_stadium_2[stadium] = max_2
  min_runs_stadium_1[stadium] = min_1
  min_runs_stadium_2[stadium] = min_2


print(avg_runs_stadium_1)
print(max_runs_stadium_1)
print(min_runs_stadium_1)
print(avg_runs_stadium_2)
print(max_runs_stadium_2)
print(min_runs_stadium_2)






```

```python

def batsmen_bowler_metrics(X):

  combined_strike_rate = []
  combined_economy_rate = []
  combined_batting_average = []
  
  all_batsmen = batsmen_strike.keys()
  all_bowlers = bowler_economy.keys()


  for index, row in X.iterrows():
    batsmen_list = row['batsmen']
    print(batsmen_list)
    bowler_list = row['bowlers']
    print(bowler_list)
    strike_rates = []
    batting_average_list = []
    for batsmen in batsmen_list:
      if batsmen not in all_batsmen:
        print("BATSMEN NOT FOUND", batsmen)
        strike_rates.append(statistics.mean(batsmen_strike.values()))
      else:
        strike_rates.append(batsmen_strike[batsmen])
      if batsmen not in batting_average.keys():
        batting_average_list.append(0)
      else:
        batting_average_list.append(batting_average[batsmen])
    combined_strike_rate.append(np.sum(strike_rates))
    combined_batting_average.append(np.sum(batting_average_list))

    economy_rates = []
    for bowler in bowler_list:
      if(bowler not in all_bowlers):
        print("BOWLER NOT FOUND", bowler)
        economy_rates.append(statistics.mean(bowler_economy.values()))
      else:
        economy_rates.append(bowler_economy[bowler])
    combined_economy_rate.append(np.sum(economy_rates))

  X['combined_strike_rate'] = pd.Series(combined_strike_rate)
  X['combined_batting_average'] = pd.Series(combined_batting_average)
  X['combined_economy_rate'] = pd.Series(combined_economy_rate)
```
